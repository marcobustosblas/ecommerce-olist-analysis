## üó∫Ô∏è 2. Mapa del Negocio: Entendiendo las Tablas

Para reconstruir el flujo de una venta, es vital entender la relaci√≥n entre las entidades principales:

### üìå INFORMACI√ìN B√ÅSICA: Entendiendo las Tablas del Modelo Olist

#### üéØ Objetivo
Comprender qu√© dato representa cada tabla y c√≥mo se conectan para reconstruir el proceso completo de un pedido.  
**Nota:** El dise√±o incluye **dos niveles de identificaci√≥n de cliente**: `customer_id` (transaccional) y `customer_unique_id` (cliente real).

---

#### üóÇ Tablas Principales

#### 1. `customers` (Clientes)
**¬øQu√© es?**  
Registro maestro de clientes.

**Claves importantes:**
- `customer_id`:  
  Identificador t√©cnico del cliente en una transacci√≥n espec√≠fica.  
  Un mismo cliente real puede tener m√∫ltiples `customer_id`.
- `customer_unique_id`:  
  Identificador √∫nico del cliente real.  
  Permite agrupar m√∫ltiples pedidos realizados por la misma persona, incluso si utiliz√≥ diferentes datos de registro.

**Relaciones:**
- `customer_id` conecta con `orders.customer_id`
- `customer_unique_id` se usa para an√°lisis de comportamiento, recurrencia y fidelidad

---

#### 2. `orders` (Pedidos)
**¬øQu√© es?**  
El n√∫cleo del sistema. Registra cada transacci√≥n de compra.

**Clave primaria (PK):** `order_id`  
**Relaci√≥n:** Cada pedido pertenece a un cliente (`customer_id` como FK).  
Un pedido **est√° compuesto por** uno o m√°s √≠tems (ver `order_items`).

---

#### 3. `order_items` (Detalle de √çtems)
**¬øQu√© es?**  
El desglose de cada art√≠culo dentro de un pedido.  
Si un pedido tiene 3 art√≠culos, habr√° 3 filas aqu√≠ con el **mismo `order_id`**.

**Claves:**
- PK: Compuesta ‚Üí `order_id` + `order_item_id`  
- FK: `order_id` (orders), `product_id` (products), `seller_id` (sellers)

**L√≥gica:**
- Cada fila representa un producto espec√≠fico vendido por un vendedor.
- `price` y `freight_value` son valores **unitarios**.
- Si un cliente compra dos unidades del mismo producto ‚Üí **dos filas separadas** con el mismo `product_id` pero distinto `order_item_id`.

---

#### 4. `products` (Productos)
**¬øQu√© es?**  
Cat√°logo de todos los productos a la venta.

**Clave primaria (PK):** `product_id`  
**Relaci√≥n:** Un producto puede aparecer en m√∫ltiples √≠tems (`order_items`).

---

#### 5. `payments` (Pagos)
**¬øQu√© es?**  
Historial de transacciones de pago.

**Aclaraci√≥n:**  
Un mismo pedido (`order_id`) puede tener m√∫ltiples filas si se pag√≥ en varias cuotas o con m√©todos combinados (ej: tarjeta + cup√≥n).

**Clave for√°nea (FK):** `order_id` (orders)

---

#### üöö Flujo del Pedido
1. Un **Cliente** (`customers`) inicia una **Orden** (`orders`).
2. La orden se desglosa en **√çtems** (`order_items`) de diversos **Productos** (`products`).
3. El proceso se completa con uno o varios **Pagos** (`payments`).
