```python
# === 2. Documentar el Mapa del Negocio. Entendiendo las Tablas === 
```

```python
"""
1- `orders`: tabla principal del sistema. Contiene informacion de los pedidos, cada fila es un pedido.
2- `order_items`: Contiene información específica de cada pedido, como el vendedor y producto asociado.
3- `products`: Tabla que incluye datos del producto, entre ellos, fotos, peso y largo.
4- `customers`: Contine información específica del cliente, como su id unico, ciudad y estado.
5- `payments`: Tabla que incluye datos del pago realizado, detalles como el tipo de pago y su valor.

OBJETIVO: Comprender qué dato representa cada tabla y cómo se conectan para reconstruir el proceso completo de un pedido.

1. `customers` (Clientes):
   - **¿Qué es?**: Registro maestro de clientes.
   - **Clave principal**: `customer_id` (ID único de cada cliente).
   - **Relación**: Un cliente puede realizar múltiples pedidos. Se conecta con `orders` a través del `customer_id`.

2. `orders` (Pedidos):
   - **¿Qué es?**: El núcleo del sistema. Registra cada transacción de compra.
   - **Clave principal**: `order_id` (ID único de cada pedido).
   - **Relación**: Cada pedido pertenece a un cliente (`customer_id`) y contiene uno o más ítems (`order_items`).

3. `order_items` (Ítems de los Pedidos):
   - **¿Qué es?**: El detalle de 'qué' se compró en cada pedido.
   - **Clave principal**: Compuesta (`order_id` + `product_id`). Una fila por cada producto distinto dentro de un pedido.
   - **Relación**: Pertenece a un `order_id` y referencia un `product_id`. Contiene el precio, cantidad y costo de envío.

4. `products` (Productos):
   - **¿Qué es?**: Catálogo de todos los productos a la venta.
   - **Clave principal**: `product_id` (ID único de cada producto).
   - **Relación**: Un producto puede aparecer en múltiples ítems de pedidos (`order_items`).

5. `payments` (Pagos):
   - **¿Qué es?**: Historial de transacciones de pago.
   - **Aclaración importante**: Un mismo pedido (`order_id`) puede tener múltiples filas aquí si se pagó en varias cuotas o con métodos combinados (ej: tarjeta + cupón).
   - **Relación**: Cada pago está asociado a un `order_id`.

FLUJO DEL NEGOCIO (Historia de un pedido):
Un 'cliente' (customers) realiza un 'pedido' (orders).
Ese 'pedido' se compone de uno o varios 'ítems' (order_items), donde cada ítem es un 'producto' (products) específico.
El cliente realiza uno o más 'pagos' (payments) para cubrir el total de ese pedido.
"""
```
